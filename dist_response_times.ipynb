```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Define parameters
N = 1000  # Total number of responses
prop_fast = 0.9  # Proportion of responses that are fast
prop_slow = 0.1  # Proportion of responses that are slow

mean_fast = 0.1  # Average response time for fast responses in seconds
std_dev_fast = 0.02  # Standard deviation of fast response times

mean_slow = 1.5  # Average response time for slow responses in seconds
std_dev_slow = 0.4  # Standard deviation of slow response times

# Step 2: Generate distributions
fast_responses = np.random.normal(mean_fast, std_dev_fast, int(N * prop_fast))
slow_responses = np.random.normal(mean_slow, std_dev_slow, int(N * prop_slow))

# Step 3: Combine distributions
response_times = np.concatenate((fast_responses, slow_responses))

# Step 4: Create a DataFrame and add labels
df = pd.DataFrame(response_times, columns=['ResponseTime'])
df['Type'] = ['Fast']*int(N * prop_fast) + ['Slow']*int(N * prop_slow)
df = df.sample(frac=1).reset_index(drop=True)  # Shuffle the DataFrame rows

# Plotting the dataset
plt.figure(figsize=(10, 6))
sns.histplot(df, x='ResponseTime', hue='Type', element='step', stat='density', common_norm=False)
plt.title('Distribution of Web Server Response Times')
plt.xlabel('Response Time (seconds)')
plt.ylabel('Density')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Define parameters
N = 1000  # Total number of responses
prop_fast = 0.9  # Proportion of responses that are fast
prop_slow = 0.1  # Proportion of responses that are slow

mean_fast = 0.1  # Average response time for fast responses in seconds
std_dev_fast = 0.02  # Standard deviation of fast response times

mean_slow = 1.0  # Average response time for slow responses in seconds
std_dev_slow = 0.4  # Standard deviation of slow response times

# Step 2: Generate distributions
fast_responses = np.random.normal(mean_fast, std_dev_fast, int(N * prop_fast))
slow_responses = np.random.normal(mean_slow, std_dev_slow, int(N * prop_slow))

# Step 3: Combine distributions
response_times = np.concatenate((fast_responses, slow_responses))

# Step 4: Create a DataFrame and add labels
df = pd.DataFrame(response_times, columns=['ResponseTime'])
df['Type'] = ['Fast']*int(N * prop_fast) + ['Slow']*int(N * prop_slow)
df = df.sample(frac=1).reset_index(drop=True)  # Shuffle the DataFrame rows

# Plotting the dataset using KDE
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='ResponseTime', hue='Type', fill=True, common_norm=False, palette="crest", alpha=0.5)
plt.title('Kernel Density Estimate of Web Server Response Times')
plt.xlabel('Response Time (seconds)')
plt.ylabel('Density')
plt.show()
```
